**Fungsi Keanggotaan**

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi keanggotaan Nilai Mahasiswa
def rendah(x):
    a, b, c, d = 0, 0, 60, 65
    if x <= a or x >= d:
        return 0
    elif a < x < b:
        return (x - a) / (b - a)
    elif b <= x <= c:
        return 1
    elif c < x < d:
        return (d - x) / (d - c)
    else:
        return 0


def sedang(x):
    a, b, c = 60, 70, 80
    if x <= a or x >= c:
        return 0
    elif a < x < b:
        return (x - a) / (b - a)
    elif b <= x <= c:
        return (c - x) / (c - b)
    else:
        return 0


def tinggi(x):
    a, b, c, d = 75, 85, 100, 100
    if x <= a or x >= d:
        return 0
    elif a < x < b:
        return (x - a) / (b - a)
    elif b <= x <= c:
        return 1
    elif c < x < d:
        return (d - x) / (d - c)
    else:
        return 0

# Rentang nilai x
x = np.linspace(0, 100, 500)

# Menghitung nilai keanggotaan untuk setiap x
y_rendah = [rendah(i) for i in x]
y_sedang = [sedang(i) for i in x]
y_tinggi = [tinggi(i) for i in x]


plt.plot(x, y_rendah, label='Rendah (Trapesium)', color='blue')
plt.plot(x, y_sedang, label='Sedang (Segitiga)', color='green')
plt.plot(x, y_tinggi, label='Tinggi (Trapesium)', color='red')

plt.title('Fungsi Keanggotaan Nilai Mahasiswa')
plt.xlabel('Nilai Mahasiswa')
plt.ylabel('Derajat Keanggotaan')
plt.legend()

plt.grid(True)
plt.show()
```

**Fuzzyfication**

```python
def cekNilai(x):
    nRendah, nSedang, nTinggi = 0,0,0

    if x >= 0 and x <= 60:
        nRendah = 1

    elif x > 60 and x < 70:
        nRendah = (-1*((x-65)/(65-60)))
        nSedang = ((x-60)/(70-60))

    elif x == 70:
        nSedang = 1

    elif x > 70 and x < 80:
        nSedang = (-1*((x-80)/(80-70)))
        nTinggi = ((x-75)/(80-75))

    elif x >= 80:
        nTinggi = 1

    return nRendah, nSedang, nTinggi
```

```python
def cekGaji(x):
    gRendah, gSedang, gTinggi = 0,0,0

    if x >= 0 and x <= 2.0:
        gRendah = 1

    elif x > 2.0 and x < 2.5:
        gRendah = (-1*(x-2.5)/(2.5-2.0))
        gSedang = ((x-2.0)/(2.5-2.0))

    elif x == 2.5:
        gSedang = 1

    elif x > 2.5 and x < 3.0:
        gSedang = (-1*(x-3.0)/(3.0-2.5))
        gTinggi = ((x-2.5)/(3.0-2.5))

    elif x >= 3.0:
        gTinggi = 1

    return gRendah, gSedang, gTinggi
```

**Inference**

```python
def inference(nRendah, nSedang, nTinggi, gRendah, gSedang, gTinggi):

    yes_list = []
    no_list = []

    if nTinggi != 0 and gTinggi != 0:
        no_list.append(min(nTinggi, gTinggi))

    if nTinggi != 0 and gSedang != 0:
        yes_list.append(min(nTinggi, gSedang))

    if nTinggi != 0 and gRendah != 0:
        yes_list.append(min(nTinggi, gRendah))

    if nSedang != 0 and gTinggi != 0:
        no_list.append(min(nSedang, gTinggi))

    if nSedang != 0 and gSedang != 0:
        yes_list.append(min(nSedang, gSedang))

    if nSedang != 0 and gRendah != 0:
        yes_list.append(min(nSedang, gRendah))

    if nRendah != 0 and gTinggi != 0:
        no_list.append(min(nRendah, gTinggi))

    if nRendah != 0 and gSedang != 0:
        no_list.append(min(nRendah, gSedang))

    if nRendah != 0 and gRendah != 0:
        no_list.append(min(nRendah, gRendah))

    # Agregasi maksimum
    Yes = max(yes_list) if yes_list else 0
    No = max(no_list) if no_list else 0

    return Yes, No
```

**Defuzzyfication Metode Mamdani**

```python
def defuzzification_Mamdani(yes, no):

    fuzzy_set = {
        "NO": {"range": [0, 50], "centroid": 40, "weight": no},
        "YES": {"range": [50, 100], "centroid": 90, "weight": yes}
    }


    numerator = 0
    denominator = 0

    for key, value in fuzzy_set.items():
        centroid = value["centroid"]
        weight = value["weight"]

        numerator += weight * centroid
        denominator += weight


    crisp_output = numerator / denominator if denominator != 0 else 0

    return crisp_output
```

**Proses Perhitungan *Fuzzy System* Metode Mamdani**

```python
def hitung_fuzzy(data):
    hasil_fuzzy = []

    # Iterasi setiap baris pada dataset
    for index, row in data.iterrows():
        # Mendapatkan nilai mahasiswa dan gaji orang tua dari setiap baris di dataset
        nilai = row['NilaiUjian']
        gaji = row['PenghasilanOrtu']

        # Menghitung nilai fuzzy untuk 'Nilai Mahasiswa' dan 'Gaji Orang Tua'
        nRendah, nSedang, nTinggi = cekNilai(nilai)
        gRendah, gSedang, gTinggi = cekGaji(gaji)

        # Melakukan inferensi untuk mendapatkan nilai fuzzy Yes dan No
        yes, no = inference(nRendah, nSedang, nTinggi, gRendah, gSedang, gTinggi)

        # Defuzzifikasi untuk mendapatkan nilai crisp
        crisp_output = defuzzification_Mamdani(yes, no)

        # Menentukan apakah hasilnya 'Yes' atau 'No' berdasarkan nilai crisp
        if crisp_output > 50:
            hasil_fuzzy.append('Yes')
        else:
            hasil_fuzzy.append('No')

    # Menambahkan hasil fuzzy sebagai kolom baru pada dataset
    data['Hasil Fuzzy'] = hasil_fuzzy
    return data

# Menghitung dan menambahkan kolom hasil fuzzy
data_kip_fuzzy = hitung_fuzzy(data)

# Menampilkan hasil
data_kip_fuzzy
```

**Akurasi *Fuzzy System* Metode Mamdani**

```python
# Menambahkan kolom baru untuk hasil prediksi fuzzy dalam bentuk 1 dan 0 (digunakan untuk menghitung akurasi)
data_kip_fuzzy['Prediksi Fuzzy'] = data_kip_fuzzy['Hasil Fuzzy'].apply(lambda x: 1 if x == 'Yes' else 0)

# Menghitung jumlah prediksi yang benar
prediksi_benar = (data_kip_fuzzy['Prediksi Fuzzy'] == data_kip_fuzzy['Target']).sum()

# Menghitung total data
total_data = len(data_kip_fuzzy)

# Menghitung akurasi
akurasi = (prediksi_benar / total_data) * 100

# Menampilkan hasil akurasi
print(f"Akurasi Metode Mamdani: {akurasi:.4f}%")
```

```python
def defuzzification_Sugeno(yes, no):

    # Aturan Sugeno
    rules = [
        {"weight": yes, "z": 90},  # YES output
        {"weight": no, "z": 45}    # NO output
    ]

    # Weighted average calculation
    numerator = sum(rule["weight"] * rule["z"] for rule in rules)
    denominator = sum(rule["weight"] for rule in rules)

    crisp_output = numerator / denominator if denominator != 0 else 0
    return crisp_output
```

```python
def hitung_fuzzy_sugeno(data):
    hasil_sugeno = []

    # Iterasi setiap baris pada dataset
    for index, row in data.iterrows():
        # Mendapatkan nilai mahasiswa dan gaji orang tua dari setiap baris di dataset
        nilai = row['NilaiUjian']
        gaji = row['PenghasilanOrtu']

        # Menghitung nilai fuzzy untuk 'Nilai Mahasiswa' dan 'Gaji Orang Tua'
        nRendah, nSedang, nTinggi = cekNilai(nilai)
        gRendah, gSedang, gTinggi = cekGaji(gaji)

        # Melakukan inferensi untuk mendapatkan nilai fuzzy Yes dan No
        yes, no = inference(nRendah, nSedang, nTinggi, gRendah, gSedang, gTinggi)

        # Defuzzifikasi untuk mendapatkan nilai crisp menggunakan Sugeno
        crisp_output = defuzzification_Sugeno(yes, no)

        # Menentukan apakah hasilnya 'Yes' atau 'No' berdasarkan nilai crisp
        if crisp_output > 50:
            hasil_sugeno.append('Yes')
        else:
            hasil_sugeno.append('No')

    # Menambahkan hasil Sugeno sebagai kolom baru pada dataset
    data['Hasil Sugeno'] = hasil_sugeno
    return data

# Menghitung dan menambahkan kolom hasil fuzzy menggunakan Sugeno
data_kip_sugeno = hitung_fuzzy_sugeno(data)

# Menampilkan hasil
data_kip_sugeno
```

```python
# Menambahkan kolom baru untuk hasil prediksi Sugeno dalam bentuk 1 dan 0
data_kip_sugeno['Prediksi Sugeno'] = data_kip_sugeno['Hasil Sugeno'].apply(lambda x: 1 if x == 'Yes' else 0)

# Menghitung jumlah prediksi yang benar untuk Sugeno
prediksi_benar_sugeno = (data_kip_sugeno['Prediksi Sugeno'] == data_kip_sugeno['Target']).sum()

# Menghitung total data
total_data_sugeno = len(data_kip_sugeno)

# Menghitung akurasi
akurasi_sugeno = (prediksi_benar_sugeno / total_data_sugeno) * 100

print(f"Akurasi Metode Sugeno: {akurasi_sugeno:.4f}%")
```
